```python
#Libs initialization
from sklearn import  metrics, tree 
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
#Dataset import
input_data = pd.read_csv(r'C:\Users\sorok\Desktop\ML Techniques\Tasks\Task #1\bioresponse.csv', sep = ',')
```

```python
input_data
```

```python
#Splitting the data to test and train
train_data, test_data, train_labels, test_labels = train_test_split(input_data.drop(['Activity'], axis=1),
                                                                    input_data.Activity,
                                                                    test_size = 0.2,
                                                                    random_state = 1)
```

```python
#Decision tree classifier for the depth = 2
dt_clf_depth_2 = tree.DecisionTreeClassifier(random_state=1, max_depth = 2)
dt_clf_depth_2.fit(train_data, train_labels)
predictions = dt_clf_depth_2.predict(test_data)

#Metrics
print('accuracy = ', metrics.accuracy_score(test_labels, predictions))
print('precision = ', metrics.precision_score(test_labels, predictions))
print('recall = ', metrics.recall_score(test_labels, predictions))
print('roc_auc = ', metrics.roc_auc_score(test_labels, predictions))
print('f1_score = ', metrics.f1_score(test_labels, predictions))
print('log_loss = ', metrics.log_loss(test_labels, predictions))

#Precision-Recall curve
predictions_proba = dt_clf_depth_2.predict_proba(test_data)
prec, rec, thresh = metrics.precision_recall_curve(test_labels, predictions_proba[:,1])
plt.figure(figsize=(9, 6))
plt.grid()
plt.plot(thresh, prec[:-1], label="Precision")
plt.plot(thresh, rec[:-1], label="Recall")
plt.xlabel(u"Threshold",fontsize = 15)
plt.title(u'Precision-Recall curve',fontsize = 15)
plt.legend(fontsize = 15)

#ROC curve
fpr, tpr, thr = metrics.roc_curve(test_labels, predictions_proba[:,1])
plt.figure(figsize = (9,6))
plt.plot(fpr, tpr, label=u'ROC \n'+'ROC AUC = ' + str(round(metrics.roc_auc_score(test_labels, predictions),2)))
plt.grid()
plt.xlabel("false positive rate",fontsize = 15)
plt.ylabel("true positive rate",fontsize = 15)
plt.title(u"ROC curve",fontsize = 15)
plt.legend(fontsize = 15)
```

```python
#Task 4: Train a classifier who avoids Type II (False Negative) errors and calculate it's metrics.
#In other words, this model needs to have a recall > 0.95.

#Let's use a random forest classifier with the depth = 10:
rf_clf_depth_10 = RandomForestClassifier(n_estimators = 50, max_depth = 10, random_state = 1)
rf_clf_depth_10.fit(train_data, train_labels)
predictions = rf_clf_depth_10.predict(test_data)
predictions_proba = rf_clf_depth_10.predict_proba(test_data)

#Now we will move the threshhold for the classifier. On the Precision-Recall curve above, it could be seen that in order to obtain 0.95+ recall, we will need to move threshold to ~ 0.3:
prob_array = np.array(predictions_proba[:,1])
first_class = 1
zeros_class = 0
for prob in np.nditer(prob_array, op_flags = ['readwrite']):
    if prob > 0.31:             #Moving threshold here
        prob[...] = first_class
    else:
        prob[...] = zeros_class

# ~ 0.31 is the optimum threshold as it gives us 0.95+ recall.
print('accuracy = ', metrics.accuracy_score(test_labels, prob_array))
print('precision = ', metrics.precision_score(test_labels, prob_array))
print('recall = ', metrics.recall_score(test_labels, prob_array))
print('roc_auc = ', metrics.roc_auc_score(test_labels, prob_array))
print('f1_score = ', metrics.f1_score(test_labels, prob_array))
print('log_loss = ', metrics.log_loss(test_labels, prob_array))
```
